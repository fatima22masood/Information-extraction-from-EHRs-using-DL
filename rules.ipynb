```python
import re

def medical_RCI(sentence):
    sentence = sentence.lower()
    found = False

    # Sort keys by length (desc) to match multi-word phrases like "chest pain"
    sorted_keys = sorted(rules.keys(), key=lambda x: -len(x))

    matched_entities = set()

    for key in sorted_keys:
        # Use word boundaries and regex to match whole phrases
        pattern = r'\b' + re.escape(key.lower()) + r'\b'
        if re.search(pattern, sentence):
            matched_entities.add(key.lower())

    if matched_entities:
        for word in matched_entities:
            info = rules[word]
            print(f"\n🧠 '{word}' is classified as: {info['type']}")

            if info["type"] == "Disease":
                print(f"🔬 Symptoms: {info.get('symptoms', [])}")
                print(f"🧪 Tests: {info.get('tests', [])}")
                print(f"💊 Treatments: {info.get('treatments', [])}")

            elif info["type"] == "Symptom":
                print(f"🩺 Possible Diseases: {info.get('possible_diseases', [])}")
                print(f"🧪 Recommended Tests: {info.get('tests', [])}")

            elif info["type"] == "Condition":
                print(f"🔬 Symptoms: {info.get('symptoms', [])}")
                print(f"🧪 Tests: {info.get('tests', [])}")
                print(f"💊 Treatments: {info.get('treatments', [])}")

            elif info["type"] == "Test Marker":
                print(f"🧪 Related Tests: {info.get('related_tests', [])}")
                print(f"🔍 Suggests: {info.get('suggests', [])}")

            else:
                print(f"ℹ️ Related Info: {info}")
    else:
        print("❌ No medical entities found in the sentence.")


```
